```python
dataset = "EEG"
task_id = 2
data_params = get_dataset_class(dataset)()
hparams = get_hparams_class(dataset)()
hparams = {**hparams.alg_hparams["DSAN"], **hparams.train_params}
```

```python
train_dataloader = data_generator('./ADATIME_data/' + dataset, task_id, data_params, hparams, "train")
test_dataloader = data_generator('./ADATIME_data/'+ dataset, task_id, data_params, hparams, "test")
```

```python
class ConvVAE(nn.Module):
    def __init__(self, input_size, latent_size, sequence_length):
        super(ConvVAE, self).__init__()

        # Encoder
        self.encoder = nn.Sequential(
            nn.Conv1d(in_channels=input_size, out_channels=64, kernel_size=3, stride=1, padding=1),
            #nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=2, stride=2),
            nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3, stride=1, padding=1),
            #nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=2, stride=2),
            nn.Flatten(),
            nn.Linear(128 * (sequence_length // 4), latent_size * 2)  # *2 for mean and variance
        )

        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(latent_size, 128 * (sequence_length // 4)),
            nn.ReLU(),
            nn.Unflatten(1, (128, sequence_length // 4)),
            nn.ConvTranspose1d(in_channels=128, out_channels=64, kernel_size=4, stride=2, padding=1),
            #nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.ConvTranspose1d(in_channels=64, out_channels=input_size, kernel_size=4, stride=2, padding=1),
            nn.Sigmoid()  # Assuming input data is normalized between 0 and 1
        )

        
    def reparameterize(self, mu, log_var):
        std = torch.exp(0.5 * log_var)
        eps = torch.randn_like(std)
        return mu + eps * std

    def forward(self, x):
        # Encode
        encoding_params = self.encoder(x)
        mu, log_var = encoding_params[:, :latent_size], encoding_params[:, latent_size:]

        # Reparameterize
        z = self.reparameterize(mu, log_var)

        # Decode
        reconstruction = self.decoder(z)

        return reconstruction, mu, log_var

# Hyperparameters
input_size = data_params.input_channels
sequence_length = data_params.sequence_len
latent_size = 2
learning_rate = 1e-3
epochs = 200

# Initialize Convolutional VAE
conv_vae = ConvVAE(input_size, latent_size, sequence_length).to("cuda")

# Loss function
criterion = lambda x1, x2 : torch.sum((x1 - x2)**2, 1).mean() #nn.BCELoss(reduction='sum')  # Binary cross-entropy loss for reconstruction
kl_divergence = lambda mu, log_var: -0.5 * torch.mean(1 + log_var - mu.pow(2) - log_var.exp())

# Optimizer
optimizer = optim.Adam(conv_vae.parameters(), lr=learning_rate)

best_loss = np.inf
# Training loop
for epoch in range(epochs):
    final_loss = 0
    for batch_data in train_dataloader:
        # Permute dimensions to [batch_size, nb_channels, sequence_length]
        x,y = batch_data
        x = x.to("cuda")
        #x = x.permute(0, 2, 1)

        # Forward pass
        reconstruction, mu, log_var = conv_vae(x)
        
        # Compute reconstruction loss and KL divergence
        reconstruction_loss = criterion(reconstruction, x)
        kl_loss = kl_divergence(mu, log_var)

        # Total loss
        total_loss = reconstruction_loss + kl_loss
        final_loss = total_loss

        # Backward pass and optimization
        optimizer.zero_grad()
        total_loss.backward()
        optimizer.step()
    final_loss = final_loss.item() / len(batch_data)
    if final_loss < best_loss:
            print(epoch+1, " : ", final_loss)
            best_loss = final_loss
            torch.save(conv_vae.state_dict(), './temp/weigths/best_model2.pth')

    # Print progress
    print(f'Epoch [{epoch+1}/{epochs}], Loss: {final_loss}')

# Save the model if needed
torch.save(conv_vae.state_dict(), './temp/weigths/conv_vae_model.pth')
```

```python
datas = []
labels = []
for batch_data in test_dataloader:
    x,y = batch_data
    x = x.to("cuda")
    z = conv_vae.encoder(x)
    mu, log_var = z[:, :latent_size], z[:, latent_size:]
    x_proj = conv_vae.reparameterize(mu, log_var).cpu()
    x_proj = x_proj.detach().numpy()
    y = y.detach().numpy()
    labels.append(y)
    datas.append(x_proj)
```

```python
points =  np.concatenate(datas)
label_idcs = np.concatenate(labels)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
scatter = ax.scatter(x=points[:, 0], y=points[:, 1], s=2.0*2, 
            c=label_idcs, cmap='tab10', alpha=0.9, zorder=2)

plt.ylim([4, -10])
plt.ylim([-5, 8])
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
```
